### 调用接口写一个聊天机器人程序


````python
import json
import urllib.request

# 下面是 图灵机器人 接口的地址
api_url = "http://openapi.tuling123.com/openapi/api/v2"

# 下面是我们写的一段程序，程序能够调用上面的接口，将要和机器人讲的话输入，就输出机器人的回答：
# 即 get_message(‘你是谁’)，可以得到结果：“我是机器人”


````

下面是我们写的一段函数，函数能够调用上面的接口，

**稍等**：这里需要给零基础的同学讲一讲函数是什么？【不管里面写的什么，即便不懂得编程的同学可以理解的说法就是：有一段程序，接受一个输入，给出一个输出】



将要和机器人讲的话输入，就输出机器人的回答：即 get_message(‘你好呀’)，可以得到结果：“你也好呀”

在正课中，我们有python的教学环节，届时您就能轻松的读懂代码了！～ 届时，老师也会对其中原理和实现作出更深入的讲解。

````python
# 你看，这就是一个函数。

def get_message(message):
    userid = 'chen9cd9c7'
    req = {
    "perception":
    {
        "inputText":
        {
            "text": message
        },

        "selfInfo":
        {
            "location":
            {
                "city": "深圳",
                "province": "广州",
                "street": "XXX"
            }
        }
    },
    "userInfo": 
    {
        "apiKey":  "",
        "userId": userid
    }
    }
    req = json.dumps(req).encode('utf8')
    http_post = urllib.request.Request(api_url, data=req, headers={'content-type': 'application/json'})
    response = urllib.request.urlopen(http_post)
    response_str = response.read().decode('utf8')
    response_dic = json.loads(response_str)
    results_text = response_dic['results'][0]['values']['text']
    return results_text
````

````python
# 我们调用上面的函数十次，就可以和机器人说上十句话：

for i in range(10): # 代表循环十次
    text_input = input('我：')
    print('                       小机器人：', get_message(text_input))
````

这样，只要所有的句子结构都储存起来，所有的词语和词性都储存起来，就是解决自然语言问题的基础了。



````python
grammar = """
句子 => 名词短语 动词短语 
名词短语 => 冠词 形容词* 名词
形容词* => null | 形容词 形容词*
动词短语 => 动词 名词短语
冠词 =>  一个 | 这个
名词 =>   男人 | 女人 |  篮球 | 桌子 | 小猫 
动词 => 喜欢 | 看见
形容词 =>   蓝色的 |  好看的 | 小小的 |  年轻的 
"""

# 如何通过语法生成和解析句子，是更加复杂的过程，咱们的下节课将会涉及。
# 鉴于一些同学需要补足编程基础，在正式课程中才会有详细的讲解。
````

基于生成的对话机器人也可以有很多形式。

也可以 **根据概率生成**：
【例如有一种概率模型叫做seq2seq模型】【有什么问题？】



根据概率生成 需要有非常大量的真实的且无偏的数据，这对模型和计算能力都有非常高的要求。

````python
import random

grammar = """
句子 => 名词短语 动词短语 
名词短语 => 冠词 形容词* 名词
形容词* => null | 形容词 形容词*
动词短语 => 动词 名词短语
冠词 =>  一个 | 这个
名词 =>   男人 | 女人 |  篮球 | 桌子 | 小猫 
动词 => 喜欢 | 看见
形容词 =>   蓝色的 |  好看的 | 小小的 |  年轻的 
"""

def parser(string_rule, gen='=>', split='|'): 
    grammar_dict = {}
    l = []
    for line in string_rule.split("\n"):
        if not line: continue
        stmt,expr = line.split(gen)
        stmt = stmt.strip()
        expr = [e.split() for e in expr.split(split)]
        grammar_dict[stmt] = expr
    return grammar_dict

def generate_from_grammar(grammar, target, gen='=>', split='|'):
    def generate(target, grammar_dict):
        if target not in grammar_dict: return target
        expr = random.choice(grammar_dict[target])
        return ''.join([generate(e,grammar_dict) for e in expr if e != 'null'])
    return generate(target, parser(grammar, gen=gen, split=split))

generate_from_grammar(grammar, '句子', gen='=>')
````

所以，机器人对语言的生成，始终在依据 **规则** 或 **概率** 作出行为。  

一个侧向规则，一个侧向语义，二者结合是常用的方法
